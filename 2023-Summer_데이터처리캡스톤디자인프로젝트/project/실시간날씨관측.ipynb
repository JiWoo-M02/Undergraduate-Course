```python
import requests
import json

class Weather:
    def __init__(self):
        self.locationCode = 0
        self.date = ""
        self.time = ""
        self.PTY = 0.0
        self.REH = 0.0
        self.RN1 = 0.0
        self.T1H = 0.0
        self.UUU = 0.0
        self.VVV = 0.0
        self.VEC = 0.0
        self.WSD = 0.0

    def __str__(self):
        return (
            "locationCode = "
            + str(self.locationCode)
            + "\ndate = "
            + self.date
            + "\ntime = "
            + self.time
            + "\nPTY = "
            + str(self.PTY)
            + "\nREH = "
            + str(self.REH)
            + "\nRN1 = "
            + str(self.RN1)
            + "\nT1H = "
            + str(self.T1H)
            + "\nUUU = "
            + str(self.UUU)
            + "\nVEC = "
            + str(self.VEC)
            + "\nVVV = "
            + str(self.VVV)
            + "\nWSD = "
            + str(self.WSD)
        )
```

```python
def main():
    print("날씨의 실시간 데이터 관측이다 보니 날짜는 최근 1일전까지만 가능합니다.")
    weather = Weather()
    apiURL = "http://apis.data.go.kr/1360000/VilageFcstInfoService_2.0/getUltraSrtNcst"
    authKey = "lcAPXOPlUUuDvgQSWdjYPP0Y29wiY1Q3OH7JsaCzLah0PUiS1tQaE9gYn4sL%2F%2BFWPMWkcigKRLoDh21BnYBLuw%3D%3D"  
    nx = input("경도를 입력해주세요 : ")
    ny = input("위도를 입력해주세요 : ")
    baseDate = input("날짜를 입력해주세요(ex.20230708) : ")
    baseTime = input("시간을 입력해주세요(ex.0800) : ")
    dataType = "JSON"
    urlBuilder = apiURL + "?" + "serviceKey=" + authKey
    urlBuilder += "&" + "numOfRows=10"  # 숫자 표
    urlBuilder += "&" + "pageNo=1"  # 페이지 수
    urlBuilder += "&" + "dataType=" + dataType  # 받으려는 타입
    urlBuilder += "&" + "base_date=" + baseDate  # 조회하고싶은 날짜
    urlBuilder += "&" + "base_time=" + baseTime  # 조회하고싶은 시간 AM 02시부터 3시간 단위
    urlBuilder += "&" + "nx=" + nx  # 경도
    urlBuilder += "&" + "ny=" + ny  # 위도
    response = requests.get(urlBuilder)
    result = response.text
    jsonData = json.loads(result)
    parse_response = jsonData["response"]
    parse_body = parse_response["body"]
    parse_items = parse_body["items"]
    parse_item = parse_items["item"]
    print(" ")
    print("--------------------------")

    for item in parse_item:
        category = item["category"]
        value = float(item["obsrValue"])

        if category == "PTY":
            weather.PTY = value
        elif category == "REH":
            weather.REH = value
        elif category == "RN1":
            weather.RN1 = value
        elif category == "T1H":
            weather.T1H = value
        elif category == "UUU":
            weather.UUU = value
        elif category == "VEC":
            weather.VEC = value
        elif category == "VVV":
            weather.VVV = value
        elif category == "WSD":
            weather.WSD = value

    weather.date = baseDate
    weather.time = baseTime

    print(weather.date, weather.time, "의 날씨는 기온", weather.T1H, "도 이며,")
    print("1시간 강수량은", weather.RN1,"mm 이고,")
    print("습도는", weather.REH,"% 입니다.")
    print("--------------------------")

if __name__ == "__main__":
    main()
```
