```python
# 1.3 species, island, sex, flipper_length_mm 만을 불러와 데이터 프레임으로 표시하고
#     filpper_length_mm 의 컬럼을 Flipp 으로 표시하고 단위는 cm로 변경하여 표시하시오.
```

```python
# 특정 컬럼 불러오기
selected_columns = ['species', 'island', 'sex', 'flipper_length_mm']
df_new = df[selected_columns].copy()

# 컬럼 이름 변경
df_new.rename(columns={'flipper_length_mm': 'Flipp'}, inplace=True)

# cm로 단위 변환
df_new['Flipp'] = df_new['Flipp'] / 10
df_new
```
