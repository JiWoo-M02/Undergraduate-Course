```python
# 2.4 Fare에서 Outlier를 찾아서 적용하되 상한 하한을 결정할 때 값을 1.2 로 적용하고
#     이상값을 분류하고 displot()을 이용하여 히스토그램을 그려라
```

```python
# 이상치 찾기
Q1 = titanic['Fare'].quantile(0.25)
Q3 = titanic['Fare'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.2 * IQR
upper = Q3 + 1.2 * IQR

# 이상치 분류
outliers = titanic[(titanic['Fare'] < lower) | (titanic['Fare'] > upper)]

# 히스토그램 그리기
sns.displot(titanic['Fare'], kde = False)
plt.xlabel('Fare')
plt.ylabel('Count')
plt.title('Histogram of Fare')
plt.show()
```
