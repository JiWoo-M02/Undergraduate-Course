```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# 첫 번째 서브플롯에 첫 번째 막대 그래프 그리기
sns.barplot(data = df, x = 'species', y = "bill_length_mm", palette = "ch:.25", ax = axes[0])

# 두 번째 서브플롯에 두 번째 막대 그래프 그리기
sns.barplot(data = df, x = 'island', y = "bill_length_mm", palette = "ch:.25", ax = axes[1])

# 그래프 간격 조정
plt.tight_layout()

# 그림 표시
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (10, 6))

# 첫 번째 서브플롯에 첫 번째 막대 그래프 그리기
sns.histplot(data = df, x = "flipper_length_mm", color = '#1E9FAA', ax = axes[0])

# 두 번째 서브플롯에 두 번째 막대 그래프 그리기
sns.histplot(data = df, x = "body_mass_g", color = '#54D6C1', ax = axes[1])

# 그래프 간격 조정
plt.tight_layout()

# 그림 표시
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (10, 8))
sns.boxplot(data = df, palette = "mako", ax = axes[0]);
sns.boxplot(data = df, x = 'species', y = 'body_mass_g', palette = "mako", ax = axes[1]);
plt.tight_layout()
plt.show()
```
