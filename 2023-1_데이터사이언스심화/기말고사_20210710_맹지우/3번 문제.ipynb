```python
import pandas as pd

# 데이터 리스트
바둑이 = pd.DataFrame({'Length': [78, 77, 86, 84, 73, 79, 74, 85], 'Height': [24, 28, 20, 28, 21, 22, 19, 35]})
순이 = pd.DataFrame({'Length': [74, 78, 80, 72, 84, 73, 72, 86], 'Height': [25, 30, 25, 24, 36, 28, 23, 36]})
기순 = pd.DataFrame({'Length': [80, 72, 85, 76, 88, 68, 76, 82], 'Height': [25, 30, 25, 24, 36, 28, 22, 35]})

# DataFrame 출력
바둑이 = 바둑이.transpose()
순이 = 순이.transpose()
기순 = 기순.transpose()

print(바둑이)
print("")
print(순이)
print("")
print(기순)
```

```python
# 그림으로 표시
fig, ax = plt.subplots()
ax.scatter(바둑이.index, 바둑이['Length'], label='Length')
ax.scatter(바둑이.index, 바둑이['Height'], label='Height')
ax.set_xticks(바둑이.index)
ax.set_xticklabels(바둑이.index)
ax.set_xlabel('Index')
ax.set_ylabel('Value')
ax.set_title('Doggy')
ax.legend()

plt.show()
```
