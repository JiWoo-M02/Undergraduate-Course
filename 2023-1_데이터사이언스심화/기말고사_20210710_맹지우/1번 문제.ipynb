```python
# mylist = [30, 20, 40, 70, 60, 50]를 이용하여 꺾은선 그래프와 막대 그래프를 그리는 코드를 작성하시오
```

```python
import matplotlib.pyplot as plt

mylist = [30, 20, 40, 70, 60, 50]

# 꺾은선 그래프
plt.plot(mylist)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Graph')
plt.show()

# 막대 그래프
plt.bar(range(len(mylist)), mylist)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Bar Graph')
plt.show()
```
