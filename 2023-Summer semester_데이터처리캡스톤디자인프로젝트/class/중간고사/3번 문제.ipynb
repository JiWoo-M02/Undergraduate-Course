```python
# 3.1 위 리스트를 각각의 시리즈 데이터로 바꾸고 두 데이터를 컬럼 기준으로 데이터를 병합하시오. 
#     그 결과를 출력하여 보이시오
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_1 = {'음식': ['갈비탕', '냉면', '짜장면', '파스타', '라멘', '김치찌개'],
          '종류': ['한식', '한식', '중식', '양식', '일식', '한식']}
data_2 = {'음식': ['갈비탕', '냉면', '짜장면', '파스타', '라멘', '김치찌개'],
          '지역': ['서울', '부산', '제주', '제주', '광주', '광주']}

# 데이터프레임 생성
df_1 = pd.DataFrame(data_1)
df_2 = pd.DataFrame(data_2)

# 시리즈 데이터로 변환
food_series = df_1['음식']
type_series = df_1['종류']
region_series = df_2['지역']
```

```python
df_food = pd.DataFrame(food_series)
df_type = pd.DataFrame(type_series)
df_region = pd.DataFrame(region_series)

# 데이터 병합
merged_df = pd.concat([df_food, df_type, df_region], axis=1)
merged_df
```

```python
# 3.4 주식의 가격 중에서 close 가격이 200 이하인 데이터를 추출하시오. 그리고 close 와 
#     volume을 곱한 값을 새로운 컬럼을 만들어 추가하시오
```

```python
# close 가격이 200 이하인 데이터
data_200 = data_stock.loc[data_stock['close'] <= 200]
data_200
```

```python
# close와 volume을 곱한 값을 새로운 컬럼을 만들어 추가
data_stock['close_volume'] = data_stock['close'] * data_stock['volume']
data_stock
```

```python
# 3.5 3.4에서 구한 데이터로 매일의 변화량 그래프를 그려보시오.
```

```python
# 날짜를 datetime 자료형으로 변환
data_stock['date'] = pd.to_datetime(data_stock['date'])

# 변화량 계산
data_stock['change'] = data_stock['close'].diff()

# 그래프 그리기
plt.plot(data_stock['date'], data_stock['change'])
plt.xlabel('Date')
plt.ylabel('Change')
plt.title('Daily Change')
plt.show()
```
