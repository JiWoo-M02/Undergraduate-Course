```python
# 3) 결측치 (#missing)
```

```python
missing_values = summary.isnull().sum()
summ = pd.DataFrame(missing_values, columns = ['#missing'])
summ
```

```python
# 4) 결측치 비율 (%missing)
```

```python
missing_percentage = (summary.isnull().mean()) * 100
summ = pd.DataFrame(missing_percentage, columns = ['%missing'])
summ
```

```python
# 3) summ[‘min’], summ[‘max’]를 컬럼으로 표현하시오  
# 4) 구현된 함수를 실행함으로서 데이터 summary를 나타내시오.
```

```python
df = summary.describe(include = 'all').transpose()

summ['min'] = df['min']
summ['max'] = df['max']

summ['first value'] = df.loc[:, '25%']
summ['second value'] = df.loc[:, '50%']
summ['third value'] = df.loc[:, '75%']

summ
```
