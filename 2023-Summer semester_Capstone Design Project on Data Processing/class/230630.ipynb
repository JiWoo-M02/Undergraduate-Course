```python
def outlier_iqr(data, column): 
 
    global lower, upper    
    
    # 4분위 수 기준 지정  
    q1, q3 = np.quantile(data[column], 0.25), np.quantile(data[column], 0.75)          
    
    iqr = q3 - q1                                  # IQR 계산    
    cut_off = iqr * 1.5                            # 기준점 계산
    lower, upper = q1 - cut_off, q3 + cut_off      # lower와 upper구분값 구하기
    
    print('IQR은',iqr, '이다.')     
    print('lower 기준값은', lower, '이다.')     
    print('upper 기준값은', upper, '이다.')    
    
    data1 = data[data[column] > upper]             # 1사 분위 데이터 저장    
    data2 = data[data[column] < lower]             # 4사 분위 데이터 저장       
    
    return print('총 이상값 개수는', data1.shape[0] + data2.shape[0], '이다.')
```
