```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_percent'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
tips_byday = tips.groupby(['day', 'smoker'])
```

```python
tips_byday_percent = tips_byday['tip_percent']
tips_byday_percent.agg('mean')
```

```python
tips_byday_percent.agg(['mean', 'std', 'sum'])
```

```python
tips_byday_percent.agg([('팀_평균', 'mean'), ('팀_표준편차', 'std')])
```

```python
function_list = ['count', 'mean', 'max']
tips_byday['tip_percent', 'total_bill'].agg(function_list)
```

```python
function_tuple = [('팀_평균', 'mean'), ('팀_표준편차', 'std')]
tips_byday['tip_percent', 'total_bill'].agg(function_tuple)
```

```python
tips_byday.agg({'tip' : 'max', 'size' : 'sum'})
```

```python
tips_byday.agg({'tip_percent' : ['min', 'max', 'mean', 'std'], 'size' : 'sum'})
```

```python
tips.groupby(['day', 'smoker']).mean()
```

```python
tips.groupby(['day', 'smoker'], as_index = False).mean()
```

```python
tips.groupby(['day', 'smoker']).mean().reset_index()
```

```python
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()
```

```python
def top(data, n = 5, column = 'body_mass_g'):
    return data.sort_values(by = column)[-n:] # n이면 body_mass_g가 작은것부터 모두 나오고/ -n일때는 가장 큰 수에서 시작해서 나옴
```

```python
top(penguins, n = 8)
```

```python
penguins.groupby('sex').apply(top)
```

```python
penguins.groupby(['sex', 'species']).apply(top, n = 1, column = 'body_mass_g')
```

```python
penguins.groupby(['sex', 'species']).apply(top, n = 1, column = 'bill_length_mm')
```

```python
penguins.groupby('sex')['body_mass_g'].describe()
```

```python
penguins.groupby('sex')['body_mass_g'].describe().unstack('sex')
```

```python
penguins.groupby('sex')['body_mass_g', 'flipper_length_mm'].apply(lambda x: x.describe())
```

```python
penguins['body_mass_g_bin'] = pd.cut(penguins['body_mass_g'], bins = 5)
penguins.head()
```

```python
penguins.body_mass_g_bin.value_counts()
```

```python
penguins['flipper_length_mm_bin'] = penguins.groupby('body_mass_g_bin').flipper_length_mm.apply(pd.cut, bins = 2)
penguins.head()
```

```python
penguins.flipper_length_mm_bin.value_counts()
```

```python
penguins['bill_length_mm_bin_by_island'] = penguins.groupby('island').bill_length_mm.apply(pd.cut, bins = 2)
penguins[['island', 'bill_length_mm', 'bill_length_mm_bin_by_island']].head(10)
```

```python
penguins[['island', 'bill_length_mm', 'bill_length_mm_bin_by_island']].tail()
```

```python
penguins.pivot_table(index = ['species', 'sex'])
```

```python
penguins.pivot_table(values = ['body_mass_g', 'flipper_length_mm'],
                      index = ['species', 'island'],
                    columns = 'sex')
```

```python
penguins.pivot_table(values = ['body_mass_g', 'flipper_length_mm'],
                      index = ['species', 'island'],
                    columns = 'sex',
                    margins = True)
```

```python
penguins.pivot_table(values = 'body_mass_g', 
                      index = ['species', 'sex'],
                    columns = 'island',
                    margins = True,
                    aggfunc = len,
                    fill_value = 0)
```

```python
penguins.pivot_table(values = ['body_mass_g', 'flipper_length_mm'], 
                      index = ['species', 'island'],
                    columns = 'sex',
                    aggfunc = {'body_mass_g' : np.mean, 'flipper_length_mm' : [min, max, np.mean]},
                    fill_value = 0)
```

```python
pd.crosstab(index = penguins.species,
           columns = penguins.island,
           margins = True)
```

```python
pd.crosstab(index = [penguins.species, penguins.sex],
           columns = penguins.island,
           margins = True)
```

```python
pd.crosstab(index = [penguins.species, penguins.sex],
           columns = penguins.island,
           margins = True,
           dropna = False)
```

```python
data_air = pd.read_csv('C:/Users/user/Documents/하계계절학기/source/datasets/AirQualityUCI.csv', sep = ';')
data_air = data_air[['Date', 'Time', 'T', 'RH', 'AH']]
```

```python
data_air.dtypes
```

```python
data_air[['T', 'RH', 'AH']].astype(float)
```

```python
data_air.head()
```

```python
data_air['T'] = data_air['T'].str.replace(',', '.').astype(float)
data_air['RH'] = data_air['RH'].str.replace(',', '.').astype(float)
data_air['AH'] = data_air['AH'].str.replace(',', '.').astype(float)

data_air.dtypes
```

```python
data_air['Date_Time'] = data_air['Date'] + ' ' + data_air['Time']
data_air.head()
```

```python
data_air = data_air.dropna()
pd.to_datetime(data_air['Date_Time'], format = '%d/%m/%Y %H.%M.%S')
```

```python
data_air['Date_Time'] = pd.to_datetime(data_air['Date_Time'], format='%d/%m/%Y %H.%M.%S')
data_air.head()
```

```python
data_air['Date_Time'].dt.hour
```

```python
data_air['hour'] = data_air['Date_Time'].dt.hour
data_air.head()
```

```python
data_air['year'] = data_air['Date_Time'].dt.year
data_air['month'] = data_air['Date_Time'].dt.month
data_air['day'] = data_air['Date_Time'].dt.day
data_air['weekday'] = data_air['Date_Time'].dt.weekday
data_air.head()
```

```python
data_air['hour_4cut'] = pd.cut(data_air['hour'], 4, labels=['night', 'morning', 'afternoon', 'evening'])
data_air.head()
```

```python
air_byhour_4cut = data_air.groupby(['weekday', 'hour_4cut'])
air_byhour_4cut.agg(['mean', 'std'])[['T', 'RH', 'AH']].head(16)
```

```python
air_bymonth = data_air.groupby(['month', 'weekday'])
air_bymonth.agg({'T' : 'max', 'RH' : 'max', 'AH' : 'max'}).head(14)
```
