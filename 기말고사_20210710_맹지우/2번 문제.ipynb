```python
# 빈도수가 높은 상위 10개의 단어를 추려내어 수직 막대그래프를 그리고 워드클라우드를 만들어 저장
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

# 텍스트 파일 읽기
with open('steve.txt', 'r', encoding='utf-8') as file:
    text = file.read()

# 단어 카운트
words = text.split()
word_counts = Counter(words)

# 상위 10개 단어 추출
top_words = word_counts.most_common(10)

# 수직 막대그래프 그리기
word_labels = [word[0] for word in top_words]
word_values = [count[1] for count in top_words]

plt.bar(word_labels, word_values)
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.title('Top 10 word frequency')
plt.xticks(rotation = 45)
plt.show()

# 워드클라우드 생성 및 저장
wordcloud = WordCloud(width = 800, height = 400, background_color = 'white').generate_from_frequencies(word_counts)

plt.figure(figsize = (10, 5))
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.title('Word Frequency WordCloud')
plt.show()

wordcloud.to_file('wordcloud.png')
```
