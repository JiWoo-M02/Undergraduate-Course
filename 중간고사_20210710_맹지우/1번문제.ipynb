```python
import pandas as pd
import matplotlib.pyplot as plt

# .csv 파일을 불러옵니다.
data = pd.read_csv("1번문제.csv", index_col=0)

# 라인플롯 차트를 그립니다.
plt.plot(data.index, data['1분기'], label='1분기')
plt.plot(data.index, data['2분기'], label='2분기')
plt.plot(data.index, data['3분기'], label='3분기')
plt.plot(data.index, data['4분기'], label='4분기')

# 축 레이블과 제목을 추가합니다.
plt.xlabel('연도')
plt.ylabel('매출액')
plt.title('분기별 매출액 추이')

# 범례를 추가합니다.
plt.legend()

# 차트를 출력합니다.
plt.show()
```
