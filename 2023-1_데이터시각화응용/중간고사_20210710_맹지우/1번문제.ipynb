```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# Actual transactions from UK retailer
df = pd.read_csv("online_retail.csv", dtype = {'CustomerID': str,'InvoiceID': str}, encoding="ISO-8859-1")
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format="%m/%d/%Y %H:%M")
```

```python
df = df.dropna()
df.head()
```

```python
# 1. Description 피처의 텍스트 정보를 확인해라
df['Description'].tolist()
```

```python
# 2. 텍스트 데이터셋에서 단어를 추출해라.
import collections

corpus = " ".join(df['Description'].tolist()).split(" ")
count = collections.Counter(corpus)
print(count)
```

```python
# 3. 추출한 단어를 이용하여 워드 클라우드 출력

import random
import pytagcloud
import webbrowser

ranked_tags = count.most_common(100)
taglist = pytagcloud.make_tags(ranked_tags, maxsize = 200)
pytagcloud.create_tag_image(taglist, '1번문제_워드클라우드.jpg', 
                            size = (900, 600),
                            rectangular = False)

from IPython.display import Image
Image(filename = '1번문제_워드클라우드.jpg')
```
