```python
import pandas as pd

data = {
    'movie': ['brother', 'brother', 'brother', 'brother', 'brother', 'brother', 'hulk', 'hulk', 'hulk', 'hulk', 'hulk', 'hulk', 'king', 'king', 'king', 'king', 'king', 'king', 'lucky', 'lucky', 'lucky', 'lucky', 'lucky', 'lucky'],
    'brand': ['daehan', 'megabox', 'cgv', 'daehan', 'megabox', 'cgv', 'daehan', 'megabox', 'cgv', 'daehan', 'megabox', 'cgv', 'daehan', 'megabox', 'cgv', 'daehan', 'megabox', 'cgv', 'daehan', 'megabox', 'cgv', 'daehan', 'megabox', 'cgv'],
    'local': ['강남', '강남', '강남', '신촌', '신촌', '신촌', '강남', '강남', '강남', '신촌', '신촌', '신촌', '강남', '강남', '강남', '신촌', '신촌', '신촌', '강남', '강남', '강남', '신촌', '신촌', '신촌'],
    'number': [15, 15, 30, 20, 30, 15, 25, 15, 15, 25, 20, 15, 30, 20, 25, 25, 25, 15, 25, 20, 25, 30, 15, 20]
}

df = pd.DataFrame(data)
df['local'] = df['local'].replace({'강남': 'Gangnam', '신촌': 'Sinchon'})
df_sum = df.groupby(['brand', 'local'], as_index=False)['number'].sum()
print("합계")
print(df_sum)
df_mean = df.groupby(['brand', 'local'], as_index=False)['number'].mean()
print("")
print("평균")
print(df_mean)
df_count = df.groupby(['brand', 'local'], as_index=False)['number'].count()
print("")
print("개수")
print(df_count)
```
